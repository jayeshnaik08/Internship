# Python program to find the factorial of a number.

```python
num=int(input('please enter integer upto which you want factorial: '))
factorial=1
if num==0:
    print('The factorial of 0 is 1')
else:
    for i in range(1,num+1):
        factorial = factorial*i
        print('The factorial of',i,'is',factorial)
        
```

# Python program to find whether a number is prime or composite

```python
number = int(input("Enter number : "))
if number == 0 or number == 1:
    print(number," is a neither prime nor composite")
else:
    for i in range(2,number):
        if(number % i == 0):
            print(number," is composite number")
            break
    else:
        print(number," is prime number")
```

# Python program to check whether a given string is palindrome or not

```python
string= 'mom'
rev_string=reversed(string)
if list(string) == list(rev_string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

# Python program to get the third side of right-angled triangle from two given sides.

```python
import numpy as np
a=int(input("Enter number side 1 : "))
b=int(input("Enter number side 2 : "))
c=np.sqrt(a**2 + b**2)
print('Third side of right angled triangle =',c)
```
